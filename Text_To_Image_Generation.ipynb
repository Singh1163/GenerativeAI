```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
```

```python
model_id1 = 'CompVis/stable-diffusion-v1-4'
model_id2 = 'stabilityai/stable-diffusion-xl-base-1.0'
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float32, use_safetensors=True)
pipe.to('cuda')
```

```python
prompt = """ generaate high quality image of A young man about 30 year with a backpack, wearing highlight color in snow capped mountain. It is snowing in mountain He has a treakking pole and poses on summit of mountain waving both hand in air"""
```

```python
prompt2 = """A young women wearing a saaree is posing in mountain with her palm touching to hes cicks, make sure to keep it cinematic and colorful"""
```

```python
image = pipe(prompt).images[0]
```
