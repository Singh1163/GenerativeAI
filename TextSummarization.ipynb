```python
import nltk
import torch
```

```python
nltk.download('punkt')
```

```python
device = "cuda" if torch.cuda.is_available() else "CPU"
device
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_ckpt = "google/pegasus-cnn_dailymail"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
```

```python
model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)
```

```python
! pip install datasets
from datasets import load_dataset
```

```python
dataset_samsum = load_dataset("samsum")
```

```python
dataset_samsum
```

```python
dataset_samsum['train']['dialogue'][1]
```

```python
dataset_samsum['train']['summary'][1]
```

```python
def convert_input_text_to_feature(input_batch):
  input_encodings = tokenizer(input_batch['dialogue'], max_length=1024, truncation=True)
  with tokenizer.as_target_tokenizer():
    target_encodings = tokenizer(input_batch['summary'], max_length=128, truncation= True)
  return {
      'input_ids' : input_encodings['input_ids'],
      'attention_mask': input_encodings['attention_mask'],
      'labels': target_encodings['input_ids']
  }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_input_text_to_feature, batched=True)
```

```python
dataset_samsum_pt['train']
```

```python
print(dataset_samsum_pt['train']['labels'][2])
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collector = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
from transformers import TrainingArguments, Trainer
```

```python
train_agrs = TrainingArguments(
    output_dir = 'pegasus', num_train_epochs=1, warmup_steps=500,
    per_device_train_batch_size=1, per_device_eval_batch_size=1,
    weight_decay=0.01, logging_steps=10,
    eval_strategy='steps', eval_steps=500, save_steps=1e6,
    gradient_accumulation_steps=16,
)
```

```python
trainer = Trainer(model=model_pegasus, args=train_agrs,
                  tokenizer=tokenizer, data_collator=seq2seq_data_collector,
                  train_dataset=dataset_samsum_pt["train"],
                  eval_dataset=dataset_samsum_pt["validation"])
```
